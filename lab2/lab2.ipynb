```python
#import libs
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
#func x^2 + 8y + 15
def func(x):
    return x[0]**2 + 8*x[1] + 15    # функція двох змінних
```

```python
#form datasets
x = np.linspace(0, 9, 10)             # відрізок 0 до 10
x0, x1 = np.meshgrid(x, x)
x0, x1 = x0.ravel(), x1.ravel()
X = [[x0[i],x1[i]] for i in range(len(x0))]
y = [func(x) for x in X]
split = int(len(X)*0.8)
x_train, y_train = X[:split], y[:split] # 0 до 8 - навчання
x_test, y_test = X[split:], y[split:]   # 8 до 10 - тестування   
```

```python
#visualizatiion
def visual(his):
    loss = his.history['loss']
    val_loss = his.history['val_loss']
    epochs = range(len(loss))
    plt.figure(figsize=(10, 6))
    plt.plot (epochs, loss, label='Training')
    plt.plot(epochs, val_loss, label='Validation')
    plt.title("Studing and cheking")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# feed forward  1 внутрішній шар з 10 нейронами
model_fForward10 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, input_shape=(2,), activation='relu'),
    tf.keras.layers.Dense(1)
])

model_fForward10.summary()
```

```python
#compile model, mean absolute error
model_fForward10.compile(optimizer=tf.keras.optimizers.SGD(
    learning_rate=tf.keras.optimizers.schedules.ExponentialDecay(0.001,
    decay_steps= 75,
    decay_rate=0.96)),
    loss='mae', metrics=['mae']
)
```

```python
history_fF10 = model_fForward10.fit(x_train, y_train, epochs=100, validation_data = (x_test, y_test))
```

```python
visual(history_fF10)
```
